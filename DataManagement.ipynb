```python
def clean_dataset(data):
    data = data.drop(['RegionID', 'SizeRank', 'RegionType'], axis=1)
    
    data['StateName'] = data['RegionName'].map(states)
    
    df = pd.melt(data, id_vars=['RegionName', 'StateName'],
                 var_name='Date', value_name='AvgPrice')

    df = df.bfill()

    nanvalues = df.isnull().sum().sum())
    if nanvalues = 0:
        df = df.to_csv('singlefamily_clean.csv', index=False)
    else:
        print('Null values present')
        break
```
