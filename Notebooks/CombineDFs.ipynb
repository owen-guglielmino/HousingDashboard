```python
import pandas as pd
```

```python
rates = pd.read_csv('mortgage.csv')
df = pd.read_csv('df_merged.csv')
```

```python
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
rates['Date'] = rates['DATE']
rates = rates.drop('DATE', axis=1)
rates
```

```python
df = df.merge(rates, on='Date', how='left')
df = df.dropna()
df
```

```python
df['MonthlyPayment'] = ((df['AvgPrice'] * 0.80) * (df['MORTGAGE30US'] / 100)) / 12
```

```python
df['MonthlyPayment'] = df['MonthlyPayment'].round(2)
df
```

```python
df = df.to_csv('df_mortgage_rates.csv')
```
